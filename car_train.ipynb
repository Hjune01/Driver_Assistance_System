```python
from google.colab import files
import os

# Ensure we are in /content before uploading
%cd /content
print(f"현재 작업 디렉토리: {os.getcwd()}") # 현재 디렉토리 확인

# --- Create the target datasets directory ---
!mkdir -p datasets
print("'/content/datasets/' 디렉토리가 생성되었습니다.")

# --- Upload the file ---
print("\n'car_model_dataset.zip' 파일을 지금 업로드해주세요:")
uploaded = files.upload()

# --- 디버그 단계 1: 업로드 후 /content/ 에 어떤 파일이 있는지 확인 ---
print("\n--- 업로드 후 /content/ 폴더 내용 ---")
!ls -l /content/
print("--------------------------------------")

# Get the filename as uploaded and potentially renamed by Colab
uploaded_filename_on_disk = None
for original_fn in uploaded.keys():
    # 먼저 원래 이름으로 파일이 존재하는지 확인
    if os.path.exists(f"/content/{original_fn}"):
        uploaded_filename_on_disk = original_fn
        break
    # Colab이 이름을 변경했을 경우 (예: ' (1).zip') 확인
    base, ext = os.path.splitext(original_fn)
    for i in range(1, 10): # 최대 9번까지의 (숫자) 이름 변경 시도
        renamed_fn = f"{base} ({i}){ext}"
        if os.path.exists(f"/content/{renamed_fn}"):
            uploaded_filename_on_disk = renamed_fn
            break
    if uploaded_filename_on_disk: # 파일을 찾았으면 루프 종료
        break

if uploaded_filename_on_disk:
    print(f"\n디스크에서 감지된 업로드 파일: '{uploaded_filename_on_disk}'")

    # --- 압축 해제: 파일을 직접 /content/ 에 압축 해제 ---
    # 이렇게 하면 'car_model_dataset' 폴더가 /content/ 바로 아래에 생성됩니다.
    print(f"'{uploaded_filename_on_disk}' 파일을 /content/ 에 압축 해제 중...")
    !unzip -q "{uploaded_filename_on_disk}" -d /content/

    # --- 디버그 단계 2: 압축 해제 후 /content/ 폴더 내용 확인 ---
    print("\n--- 압축 해제 후 /content/ 폴더 내용 ---")
    !ls -l /content/
    print("------------------------------------------")

    # --- 압축 해제된 폴더를 /content/datasets/ 로 이동 ---
    # 이 부분은 zip 파일 안에 'car_model_dataset'이라는 최상위 폴더가 직접 들어있다고 가정합니다.
    if os.path.exists("/content/car_model_dataset"):
        print("'/content/car_model_dataset' 폴더를 '/content/datasets/' 로 이동 중...")
        !mv /content/car_model_dataset /content/datasets/
        print("폴더 이동 완료.")
    else:
        print("경고: 압축 해제 후 '/content/car_model_dataset' 폴더를 찾을 수 없습니다. zip 파일의 내부 구조를 확인해주세요.")
        print("압축 파일이 'car_model_dataset/' 폴더 대신 'images/'와 'labels/'를 바로 포함할 수도 있습니다.")

        # 대안: 만약 zip 파일이 'images/'와 'labels/' 폴더를 직접 포함하고 있다면, 이들을 이동시킵니다.
        if os.path.exists("/content/images") and os.path.exists("/content/labels"):
            print("'/content/'에서 'images'와 'labels'를 직접 감지했습니다. 이들을 '/content/datasets/car_model_dataset/'로 이동 중...")
            !mkdir -p /content/datasets/car_model_dataset/ # 중간 폴더 생성
            !mv /content/images /content/datasets/car_model_dataset/
            !mv /content/labels /content/datasets/car_model_dataset/
            print("대안 이동 완료.")
        else:
            print("압축 해제 후 '/content/'에서 'car_model_dataset', 'images', 'labels' 폴더 중 어떤 것도 찾을 수 없습니다.")


else:
    print("오류: 업로드 후 인식되거나 찾을 수 있는 .zip 파일이 없습니다.")


# --- 최종 데이터셋 폴더 구조 확인 ---
print("\n최종 데이터셋 폴더 구조 확인:")
!ls -R /content/datasets/
```

```python
import os

# 현재 작업 디렉토리를 yolov5/ 에셋 디렉토리로 변경
# %cd /content/yolov5/

# datasets 폴더가 /content/datasets/car_model_dataset 에 있다고 가정
# 이 경로는 data.yaml 파일에서 사용됩니다.
DATASET_PATH = '/content/datasets/car_model_dataset'

# 클래스 개수 (detection할 객체의 종류 수)
# 예: 'car' 하나만 있다면 1, 'car', 'truck' 두개라면 2
NUM_CLASSES = 1 # 여기에 실제 클래스 개수를 입력하세요 (예: 1)

# 클래스 이름 (실제 사용한 클래스 이름을 리스트로 입력)
# 예: ['car'] 또는 ['car', 'truck']
CLASS_NAMES = ['car'] # 여기에 실제 클래스 이름을 입력하세요 (예: ['car'])

# data.yaml 파일 경로 설정
yaml_path = '/content/yolov5/data/car_model_data.yaml' # yolov5 폴더 안에 새로운 yaml 파일 생성

# data.yaml 파일 내용 생성
yaml_content = f"""
# Train/val/test sets as 1) dir: path/to/imgs, 2) file: path/to/imgs.txt, or 3) list: [path/to/imgs1, path/to/imgs2]
path: {DATASET_PATH}  # dataset root dir
train: images/train  # train images (relative to 'path') 128 images
val: images/val  # val images (relative to 'path') 128 images
# test: images/test # test images (optional)

# Classes
nc: {NUM_CLASSES}  # number of classes
names: {CLASS_NAMES} # class names list
"""

# data.yaml 파일 쓰기
with open(yaml_path, 'w') as f:
    f.write(yaml_content)

print(f"'{yaml_path}' 파일이 성공적으로 생성되었습니다.")
print("--- car_model_data.yaml 내용 ---")
!cat {yaml_path}
print("------------------------------")

# 파일이 제대로 생성되었는지 확인
if os.path.exists(yaml_path):
    print("\n'car_model_data.yaml' 파일이 성공적으로 준비되었습니다. 다음 단계로 진행할 수 있습니다.")
else:
    print("\n오류: 'car_model_data.yaml' 파일 생성에 실패했습니다. 코드를 확인해주세요.")
```

```python
# 현재 작업 디렉토리가 /content/yolov5/ 인지 확인
# 만약 아니라면, 아래 명령어를 먼저 실행해주세요:
%cd /content/yolov5

print("YOLOv5 모델 학습을 시작합니다...")

# 학습 명령어 실행
# --img: 학습에 사용할 이미지 크기 (보통 640x640)
# --batch-size: 배치 사이즈. GPU 메모리에 따라 조절 (너무 크면 OOM 오류 발생)
# --epochs: 학습할 에폭(Epoch) 수. 데이터셋 크기에 따라 적절히 조절
# --data: 우리가 생성한 data.yaml 파일 경로
# --cfg: 사용할 모델 설정 파일 (예: yolov5s.yaml, yolov5m.yaml 등)
# --weights: 초기 가중치 파일 (사전 학습된 모델 사용)
# --name: 이 학습 세션의 이름 (결과 저장 폴더 이름이 됨)

!python train.py --img 640 --batch-size 16 --epochs 50 --data data/car_model_data.yaml --cfg models/yolov5s.yaml --weights yolov5s.pt --name car_model_training_results

print("\nYOLOv5 학습 명령어가 실행되었습니다. 학습 완료까지 시간이 소요됩니다.")
print("진행 상황은 위 출력에서 확인하실 수 있습니다.")
print("학습 결과는 '/content/yolov5/runs/train/car_model_training_results' 폴더에 저장됩니다.")
```

```python
import os

# 현재 작업 디렉토리가 /content/yolov5/ 인지 확인
# 만약 아니라면, 아래 명령어를 먼저 실행해주세요:
%cd /content/yolov5

print("PyTorch 모델 (best.pt)을 ONNX 형식으로 변환합니다...")

# export.py 스크립트 실행
# --weights: 변환할 학습된 모델 가중치 파일 (runs/train/YOUR_TRAINING_NAME/weights/best.pt)
#            이전 학습에서 --name 인자에 지정했던 폴더 이름을 확인하여 경로를 정확히 맞춰주세요.
# --include onnx: ONNX 형식으로 내보내기 지정
# --imgsz: 모델 입력 이미지 크기 (학습 시 사용한 크기와 동일하게 640을 권장)
# --batch-size: 추론 시 일반적으로 배치 사이즈는 1을 사용합니다.
# --simplify: ONNX 모델을 단순화하여 더 효율적으로 만듭니다. (성능 향상)
# --opset: ONNX Operator Set 버전 (최신 버전 사용 권장, 보통 12 또는 13)

# 예시 경로: 'runs/train/car_model_training_results/weights/best.pt'
# 만약 학습 시 --name을 다르게 설정했다면 'car_model_training_results' 부분을 수정해주세요.
!python export.py --weights runs/train/car_model_training_results/weights/best.pt --include onnx --imgsz 640 --batch-size 1 --simplify --opset 12

print("\nONNX 변환 명령어가 실행되었습니다. 변환이 완료되면 '.onnx' 파일이 생성됩니다.")
print("생성된 ONNX 파일은 'runs/train/car_model_training_results/weights/best.onnx' 경로에 있을 것입니다.")

# 파일이 제대로 생성되었는지 확인
onnx_model_path = 'runs/train/car_model_training_results/weights/best.onnx'
if os.path.exists(onnx_model_path):
    print(f"\n'{onnx_model_path}' 파일이 성공적으로 생성되었습니다!")
    print(f"이제 이 '{onnx_model_path}' 파일을 컴퓨터로 다운로드 받으세요.")
    print("다운로드 방법: Colab 왼쪽 파일 탐색기에서 해당 파일을 찾아 마우스 오른쪽 버튼 클릭 -> 다운로드")
else:
    print(f"\n오류: '{onnx_model_path}' 파일 생성에 실패했습니다. export.py 출력을 확인해주세요.")
```

```python
# Colab에서 실행
%cd /content/yolov5

print("ONNX 모델을 INT8 양자화된 형식으로 변환합니다...")

# --int8 플래그 사용: INT8 양자화 활성화
# --dynamic 플래그 추가 (선택 사항이지만 CPU에서는 일반적으로 동적 양자화가 유용합니다)
# --weights: 변환할 학습된 모델 가중치 파일 (이전 학습에서 사용한 경로 그대로)
# --include onnx: ONNX 형식으로 내보내기
# --imgsz: 모델 입력 이미지 크기 (640 유지)
# --batch-size: 1 (추론 시 일반적으로)
# --simplify: ONNX 모델 단순화
# --opset: ONNX Operator Set 버전 (기존과 동일)

# 예시 경로: 'runs/train/car_model_training_results/weights/best.pt'
!python export.py --weights runs/train/car_model_training_results/weights/best.pt --include onnx --imgsz 640 --batch-size 1 --simplify --opset 12 --int8 --dynamic

print("\nINT8 양자화된 ONNX 변환 명령어가 실행되었습니다.")
print("생성된 ONNX 파일은 'runs/train/car_model_training_results/weights/best.onnx' (기존 파일 덮어쓰기)에 있을 것입니다.")
print("이 양자화된 파일을 라즈베리 파이로 다시 옮겨서 사용해 보세요.")
```
